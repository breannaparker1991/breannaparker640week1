```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data to dataframe
df_hcw=pd.read_excel('hotdog-contest-winners.xlsm')
df_hcw
```

```python
df_hcw_h = df_hcw.head()
df_hcw_h
```

```python
#pie chart
plot = df_hcw.plot.pie(y='Dogs eaten', labels = df_hcw['Winner'], figsize=(5, 5))
```

```python
#just used the head since there's a lot of data. 
plot = df_hcw_h.plot.pie(y='Dogs eaten', labels = df_hcw['Winner'], figsize=(5, 5))
```

```python
#bar chart of hot dog winners with count of dogs eaten
ax = df_hcw.plot.bar(x='Winner', y='Dogs eaten', rot=0)
ax
```

```python
#import data to dataframe
df_hp=pd.read_excel('hotdog-places.xlsm')
df_hp.index += 1
df_hp['index'] = df_hp.index
df_hp
```

```python
#donut chart
labels = ["1st place", "2nd place", "3rd place"]

colors = ['#FF0000', '#0000FF', '#FFFF00', 
          '#ADFF2F', '#FFA500']

explode = (0.05, 0.05, 0.05)

plt.pie(df_hp[2002], colors=colors, labels=labels,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
 
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
fig.gca().add_artist(centre_circle)
 
plt.title('Percentage hot dogs eaten in 2002')
 
plt.show()
```
